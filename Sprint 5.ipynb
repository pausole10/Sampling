```python
# Creo una funció per introduir directament el dataframe i el salt entre files perquè sigui molt més ràpid i automàtic. 

import numpy as np
def systematic_sampling(df, step):
 
    indexes = np.arange(0, len(df), step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample

systematic_sample = systematic_sampling(df, 3)
 

systematic_sample.head(10)
```

```python
# Només vull dos valors de la columna home_team_name. Els agafo i em retorna el nombre que hi ha de cadascun.

df2 = df[(df['home_team_name'] == 'Manchester United') | (df['home_team_name'] == 'Manchester City')]
df2
```

```python
df2['home_team_name'].unique()

```

```python
# Agafo només les columnes amb valors numèrics.

df3= df2[['timestamp', 'attendance', 'home_team_name', 'Game Week', 'Pre-Match PPG (Home)', 'Pre-Match PPG (Away)', 
  'home_ppg', 'away_ppg', 'home_team_goal_count' ]]

df3
```

```python
# Dono valors de 0 a 'Manchester United' i valor de 1 per a 'Manchester City' per tal de poder treballar amb la variable
# binària i que es pugui treballar amb la funcio SMOTE

df3['home_team_name'] = [0 if x == 'Manchester United' else 1 for x in df3['home_team_name']]
df3
```

```python
# Miro quantes unitats hi ha de cada valor definit prèviament

df3['home_team_name'].value_counts()
```

```python
#Descripció de les dades de l'últim dataframe creat

df3['home_team_name'].describe()
```

```python
# Utilitzo SMOTe per fer un resample i assegurarme que no existeix cap descompensació de les dades i no estan esbiaixades
# per cap costat.

smote = SMOTE(sampling_strategy = 'minority')
X_sm, Y_sm = smote.fit_resample(df3,df3['home_team_name'])
X_sm
```
